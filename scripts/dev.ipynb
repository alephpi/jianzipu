```python
_, paragraph = parse_file("../../../曲谱/酒狂_简写.txt")
```

```python
canvases = []
for token in paragraph:
    try:
        note = parse(token, form="abbr")
        canvas = note.draw()
        canvas.attribs["width"] = 100
        canvas.attribs["height"] = 100
        canvas.attribs["viewBox"] = "0 0 200 200"
        canvases.append(canvas)
        # display(canvas)
    except:
        print(token)
```

```python
canvas
```

```python
parse("从「再作", "abbr").draw()
```

```python
for i, canvas in enumerate(canvases):
    string = canvas.tostring()
    with open(f"../../../site/svgs/{i}.svg", "w") as f:
        f.write(string)
```
